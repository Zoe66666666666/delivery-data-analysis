```python
import pandas as pd


df = pd.read_csv('../data/delivery_data.csv')


print(df.head())


print(df.info())


print(df.isnull().sum())


print(df.dtypes)


print(df.duplicated().sum())


print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
df.isnull().sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Columns with Missing Values")
plt.ylabel("Missing Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
df.select_dtypes(include='number').hist(figsize=(12, 8), bins=20)
plt.suptitle("Distribution of Numeric Columns")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
df['trip_creation_time'].value_counts().sort_index().plot()
plt.title("Trip Creation Time Distribution")
plt.ylabel("Trip Count")
plt.xlabel("Time")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['actual_time'], bins=30, kde=True)
plt.title("Distribution of Actual Trip Time")
plt.xlabel("Time (minutes)")
plt.show()
```
